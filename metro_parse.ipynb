```python
import pandas as pd
```

```python
url = "https://mostobook.ru"
df = pd.read_html(url,skiprows=[0])[0]
```

```python
cls = ['bridge', 'river', 'stop_mv', 'start_drawing', 'end_drawing', 'start_mv']
df.columns = cls
print(df)
```

```python
import re

def splits(x):
    if len(x) == 4:
        return [x]
    if len(x) == 8:
        return [x[0:4],x[4:]]
    return [x]

ndf = pd.DataFrame(columns=cls)
i = 0
clss = ['stop_mv', 'start_drawing', 'end_drawing', 'start_mv']

for index, row in df.iterrows():
    if len(splits(row['stop_mv'])) == 1:
        ndf = ndf.append(pd.Series(row, index=cls ), ignore_index=True)
        continue
    if len(splits(row['stop_mv'])) == 2:
        row1 = pd.Series([row['bridge'],row['river'], splits(row['stop_mv'])[0],splits(row['start_drawing'])[0],
                         splits(row['end_drawing'])[0], splits(row['start_mv'])[0]], index=cls)
        ndf = ndf.append(pd.Series(row1, index=cls ), ignore_index=True)
        row2 = pd.Series([row['bridge'],row['river'], splits(row['stop_mv'])[1],splits(row['start_drawing'])[1],
                         splits(row['end_drawing'])[1], splits(row['start_mv'])[1]], index=cls)
        ndf = ndf.append(pd.Series(row2, index=cls ), ignore_index=True)

ndf = ndf.sort_values(['bridge','stop_mv'])

print(ndf)
json = ndf.to_json(orient='table', force_ascii=False)
with open("./data/bridges.json", "w+", encoding='utf-8') as text_file:
    text_file.write(json)
```
